__1) Розв’язати задачу комівояжера для графа, заданого варіантом, використовуючи код, наведений вище.__

Заданий зважений граф: [(1,2,4), (1,3,5), (1,4,7), (1,5,8), (2,3,3), (2,4,4),
(2,5,6), (3,4,2), (3,5,3), (4,5,1)]

__2) Візуалізувати граф.__

```python
import networkx as nx
import matplotlib.pyplot as plt

# Ініціалізація графа
G = nx.Graph()

# Додавання вузлів
G.add_nodes_from(range(1, 6))

# Додавання зважених ребер
edges = [(1, 2, 4), (1, 3, 5), (1, 4, 7), (1, 5, 8),
         (2, 3, 3), (2, 4, 4), (2, 5, 6),
         (3, 4, 2), (3, 5, 3), (4, 5, 1)]

G.add_weighted_edges_from(edges)

# Знаходження мінімального остового дерева за алгоритмом Крускала
mst = nx.minimum_spanning_tree(G, algorithm='kruskal')

# Візуалізація графа та MST
pos = nx.spring_layout(G)  # позиції для всіх вузлів

plt.figure(figsize=(12, 8))

# Відображення всіх вузлів
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

# Відображення всіх ребер
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), alpha=0.3, edge_color='grey')

# Відображення MST ребер
nx.draw_networkx_edges(G, pos, edgelist=mst.edges(), width=2, edge_color='blue')

# Відображення підписів ребер
edge_labels = dict([((u, v,), d['weight']) for u, v, d in G.edges(data=True)])
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Мінімальне остове дерево (MST)')
plt.show()
```
