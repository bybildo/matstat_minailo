19. –ú–∞—î–º–æ –¥–≤—ñ –∫–æ—Ä–æ—Ç–∫—ñ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ —Å–∏–º–≤–æ–ª—ñ–≤: ¬´1234¬ª —ñ ¬´2345¬ª. –ó–Ω–∞–π—Ç–∏
–Ω–∞–π–¥–æ–≤—à—É —Å–ø—ñ–ª—å–Ω—É –ø—ñ–¥–ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ —Å–∏–º–≤–æ–ª—ñ–≤, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –∞–ª–≥–æ—Ä–∏—Ç–º
–¥–∏–Ω–∞–º—ñ—á–Ω–æ–≥–æ –ø—Ä–æ–≥—Ä–∞–º—É–≤–∞–Ω–Ω—è.

```python
#   |   | 1 | 2 | 3 | 4 |
#---#--------------------
#   # 0 | 0 | 0 | 0 | 0 |
# 2 # 0 | 0 | 1 | 1 | 1 |
# 3 # 0 | 0 | 1 | 2 | 2 |
# 4 # 0 | 0 | 1 | 2 | 3 |
# 5 # 0 | 0 | 1 | 2 | 3 |
```

```python
def longest_common_subsequence(s1, s2):
 m = len(s1)
 n = len(s2)
 # –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è –ø—Ä–æ–º—ñ–∂–Ω–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
 # dp[i][j] –±—É–¥–µ –º—ñ—Å—Ç–∏—Ç–∏ –¥–æ–≤–∂–∏–Ω—É –Ω–∞–π–±—ñ–ª—å—à–æ—ó —Å–ø—ñ–ª—å–Ω–æ—ó
 dp = [[0] * (n + 1) for _ in range(m + 1)]
 # –ó–∞–ø–æ–≤–Ω–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ –∑–Ω–∏–∑—É –≤–≥–æ—Ä—É
 for i in range(1, m + 1):
     for j in range(1, n + 1):
         if s1[i - 1] == s2[j - 1]:
             dp[i][j] = dp[i - 1][j - 1] + 1
         else:
             dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
 # –í—ñ–¥–Ω–æ–≤–ª–µ–Ω–Ω—è –Ω–∞–π–±—ñ–ª—å—à–æ—ó —Å–ø—ñ–ª—å–Ω–æ—ó –ø—ñ–¥–ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ
 lcs = []
 i, j = m, n
 while i > 0 and j > 0:
     if s1[i - 1] == s2[j - 1]:
         lcs.append(s1[i - 1])
         i -= 1
         j -= 1
     elif dp[i - 1][j] > dp[i][j - 1]:
         i -= 1
     else:
         j -= 1
 # –ü–µ—Ä–µ–≤–µ—Ä–Ω–µ–Ω–Ω—è lcs, –æ—Å–∫—ñ–ª—å–∫–∏ –º–∏ –¥–æ–¥–∞–≤–∞–ª–∏ —Å–∏–º–≤–æ–ª–∏ –∑ –∫—ñ–Ω—Ü—è
 lcs.reverse()
 return lcs
# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
s1 = "1234"
s2 = "2345"
print("–ù–∞–π–±—ñ–ª—å—à–∞ —Å–ø—ñ–ª—å–Ω–∞ –ø—ñ–¥–ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å:",
''.join(longest_common_subsequence(s1, s2)))
```
