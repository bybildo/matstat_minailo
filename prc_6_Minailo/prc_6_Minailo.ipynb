1. _Що таке граф і які головні складові його структури?_
   
Граф — це сукупність об'єктів із зв'язками між ними. 
Об'єкти розглядаються як вершини, або вузли графу, а зв'язки — як дуги, або ребр

2. _Які алгоритми використовуються для пошуку найкоротших шляхів у 
графах_

Алгоритм Дейкстри, Белмена–Форда, Флойда–Форшала

3. _Як працює алгоритм Дейкстри і які його особливості?_

Алгоритм Дейкстри — це популярний алгоритм для вирішення задач найкоротших шляхів. Кожній вершині з безлічі вершин V порівняємо мітку - мінімальна відома відстань від цієї вершини до стартової вершини a. Алгоритм працює покроково - на кожному кроці він "відвідує" одну вершину і намагається зменшувати мітки. 
Робота алгоритму завершується коли всі вершини відвідані. Алгоритм працює лише з графами, де ваги ребер не від’ємні.
Він може працювати як на напрямлених, так і на ненапрямлених графах
   
   
5. _Що таке алгоритм Белмена–Форда і коли його варто застосовувати?
   Цей алгоритм пошуку найкоротшого шляху знаходить найкоротші шляхи від однієї вершини графу до всіх інших. На відміну від алгоритму Дейкстри, алгоритм Белмена–Форда допускає ребра з негативною вагою. І саме тоді коли граф може містити ребра з негативною вагою потрібно використовувати цей алгоритм._

   75._ Як працює алгоритм Флойда–Форшала і які його переваги та недолік

Алгоритм може працювати з графами, що містять ребра з додатними або від’ємними вагами. Однак, він не може працювати з графами, що містять негативні цикли. Алгоритм Флойда–Форшала працює за принципом порівняння всіх можливих шляхів в графі між кожною парою вершин. Переваги: знаходить найкоротші шляхи між усіма парами, може працювати і з додатніми і з від'ємними вагами.и_??а.
