```python
def BinaryTree(r):
    return [r, [], []]

def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
         root.insert(1,[newBranch,t,[]])
    else:
        root.insert(1,[newBranch, [], []])
    return root

def getRootVal(root):
    return root[0]

def setRootVal(root,newVal):
    root[0] = newVal

def getLeftChild(root):
    return root[1]

def getRightChild(root):
    return root[2]

#--------------------------------------------------------------#-----------------------------------------------------------------------------------------
def deleteBranch(root, branchToDelete):                        #        У найкращому / У найгіршому
    if root == branchToDelete:                                 # c1          1       /       1
        return []                                              # c2          1       /       0
    root[1] = deleteBranch(getLeftChild(root), branchToDelete) # c3          0       /       n
    root[2] = deleteBranch(getRightChild(root), branchToDelete)# c4          0       /       n                                                              # c5          0       /       1                          
    return root                                                # c5          0       /       1
                                                               #            O(1)     /      O(n) 
#----------------------------------------------#---------------------------------------------------------------------------------------------------------
def insertRight(root, newBranch):              #        У найкращому / У найгіршому
    t = root.pop(2)                            # с1           1      /      1
    if len(t) > 1:                             # с2           1      /      1
        root.insert(2,[newBranch,[],t])        # с3           0      /      1
    else:                                      # с4           1      /      1
        root.insert(2,[newBranch,[],[]])       # с5           0      /      0
    return root                                # с6           1      /      1
                                               #             O(1)    /     O(1)
#----------------------------------------------#---------------------------------------------------------------------------------------------------------
def search(root, value):                       #        У найкращому / У найгіршому
    if getRootVal(root) == value:              # c1           1      /       1
        return True                            # c2           1      /       0
    else:                                      # c3           0      /       1
        left = getLeftChild(root)              # c4           0      /       1
        if left:                               # c5           0      /       1
            if search(left, value):            # c6           0      /       n
                return True                    # c7           0      /       0
        right = getRightChild(root)            # c8           0      /       1
        if right:                              # c9           0      /       1
            if search(right, value):           # c10          0      /       n
                return True                    # c11          0      /       0
    return False                               # c12          0      /       1
                                               #             O(1)    /     O(n)
#----------------------------------------------#---------------------------------------------------------------------------------------------------------


r = BinaryTree(3)
insertLeft(r,4)
insertLeft(r,5)
insertRight(r,6)
insertRight(r,7)
l = getLeftChild(r)
print(l)
setRootVal(l,9)
print(r)
insertLeft(l,11)
print(r)
print(getRightChild(getRightChild(r)))
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, key):
        if isinstance(key, list):
            key = tuple(key)
        elif isinstance(key, dict):
            key = frozenset(key.items())
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        self.table[index].append((key, value))

    def search(self, key):
        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return None

    def delete(self, key):
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                del self.table[index][i]
                return True
        return False

import time

def measure_time(func, *args):
    start_time = time.perf_counter()
    func(*args)
    end_time = time.perf_counter()
    return end_time - start_time

def test_hash_table():
    print("Тестування з цілими числами:")
    hash_table_int = HashTable(10)
    print("Час вставки:", measure_time(hash_table_int.insert, 5, "value"))
    print("Час пошуку:", measure_time(hash_table_int.search, 5))
    print("Час видалення:", measure_time(hash_table_int.delete, 5))

    print("\nТестування з рядками:")
    hash_table_str = HashTable(10)
    print("Час вставки:", measure_time(hash_table_str.insert, "key", "value"))
    print("Час пошуку:", measure_time(hash_table_str.search, "key"))
    print("Час видалення:", measure_time(hash_table_str.delete, "key"))

    print("\nТестування зі списками:")
    hash_table_list = HashTable(10)
    print("Час вставки:", measure_time(hash_table_list.insert, [1, 2, 3], "value"))
    print("Час пошуку:", measure_time(hash_table_list.search, [1, 2, 3]))
    print("Час видалення:", measure_time(hash_table_list.delete, [1, 2, 3]))

    print("\nТестування з словниками:")
    hash_table_dict = HashTable(10)
    print("Час вставки:", measure_time(hash_table_dict.insert, {"key": "value"}, "value"))
    print("Час пошуку:", measure_time(hash_table_dict.search, {"key": "value"}))
    print("Час видалення:", measure_time(hash_table_dict.delete, {"key": "value"}))

    class TestObject:
        def __init__(self, value):
            self.value = value

    print("\nТестування з об'єктами:")
    hash_table_obj = HashTable(10)
    obj = TestObject("test")
    print("Час вставки:", measure_time(hash_table_obj.insert, obj, "value"))
    print("Час пошуку:", measure_time(hash_table_obj.search, obj))
    print("Час видалення:", measure_time(hash_table_obj.delete, obj))

test_hash_table()
```

##### _Чим відрізняється структура бінарне дерево він бінарного дерева пошуку?_
 
Бінарне дерево це дерев’яна структура даних, в якій кожен вузол має максимум два дочірніх вузла, а у бінарного дерева пошуку для кожного вузла X, всі значення в лівому піддереві X менші за значення X, і всі значення в правому піддереві X більші за значення X.
 
##### _Чим відрізняється структура бінарне дерево від бінарної купи?_

Коли у бінарному дереві кожен вузол має не більше двох дочірніх вузлів, бінарна купа завжди є повним бінарним деревом.

##### _Які існують типи дерев? Опишіть їхні основні характеристики та переваги._
 
Бінарне дерево: Це дерево, в якому кожен вузол має не більше двох дочірніх вузлів, зазвичай визначених як лівий та правий вузли. Бінарні дерева часто використовуються для реалізації бінарних дерев пошуку та бінарних куп.

Бінарне дерево пошуку (BST): Це бінарне дерево, в якому всі вузли відповідають певному впорядкуванню: всі вузли в лівому піддереві менші за корінь, а всі вузли в правому піддереві більші за корінь. BST дозволяє ефективно здійснювати операції пошуку, вставки та видалення.


Бінарна купа: Це повне бінарне дерево (або майже повне), в якому значення вузла є більшим (або меншим) за значення його дочірніх вузлів. Це властивість робить бінарну купу корисною для реалізації пріоритетних черг.

##### _Опишіть методи вирішення колізій в хеш-таблицях. Які їхні переваги та недоліки?_
 
Колізії в хеш-таблицях виникають, коли два різних ключа мають однаковий геш-код, тобто вони вказують на одне й те ж місце в таблиці. Існують різні методи вирішення колізій:

Лінійне зондування: Якщо виникає колізія, просто переходимо до наступного слоту в таблиці і перевіряємо, чи вільний він. Якщо так, вставляємо елемент туди. Якщо ні, продовжуємо перевіряти наступні слоти.

Недоліки: Може призвести до “кластеризації”, коли групи вільних або зайнятих слотів збільшуються, що погіршує ефективність.


Двохетапне хешування: Якщо виникає колізія, використовується друга геш-функція для визначення кроку, на який ми перейдемо до наступного слоту.

Переваги: Зменшує ймовірність кластеризації порівняно з лінійним зондуванням.

Недоліки: Більш складний для реалізації, оскільки потребує двох геш-функцій.
